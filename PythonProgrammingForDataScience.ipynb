# **Iterators and Generators**

# **Creating and Using Iterators**
An iterator is an object that contains a sequence of elements and can be iterated (looped) one element at a time. Iterators are implemented using the __iter__() and __next__() methods.

````python
# Creating an iterator from a list
my_list = [1, 2, 3, 4]
iterator = iter(my_list)  # Get an iterator from the list

# Access elements using next()
first_element = next(iterator)  # 1
second_element = next(iterator)  # 2
````

````python
# Custom iterator class
class MyIterator:
    def __init__(self, max_value):  # This initializes the iterator object with max_value (the upper limit) and current (starting value, set to 0).
        self.max_value = max_value
        self.current = 0

    def __iter__(self):  # It must have an iter() method that returns the iterator object itself.
        return self

    def __next__(self):  # It must have a next() method that returns the next value in the sequence or raises StopIteration to indicate the end.
        if self.current < self.max_value:
            result = self.current
            self.current += 1
            return result
        else:
            raise StopIteration

# Using the custom iterator
for num in MyIterator(5):
    print(num)  # Outputs: 0, 1, 2, 3, 4
````

* An iterator is an object with \__ iter \__() (returns the iterator) and \__ next \__() (returns the next item) methods.
* Use iter() to get an iterator from an iterable like a list, tuple, or string.
* Use next() to get the next item from an iterator, and StopIteration is raised when no more items are available.

# **Generator Functions and Expressions**
Generators are functions that yield items one at a time instead of returning them all at once. They are memory-efficient since they generate values on demand.

````python
# Generator function using yield
def my_generator():
    yield 1
    yield 2
    yield 3

# Using the generator
for value in my_generator():
    print(value)  # Outputs: 1, 2, 3
````

````python
# Infinite generator
def infinite_counter():
    num = 0
    while True:
        yield num
        num += 1

# Access the first 5 values from the infinite generator
counter = infinite_counter()
for _ in range(5):
    print(next(counter))  # Outputs: 0, 1, 2, 3, 4
````

* Generator functions use the yield keyword instead of return to produce values one at a time.
* Generators maintain their state between yields, allowing the computation to be resumed.
* Generators are memory-efficient as they only produce one value at a time, unlike lists that store all values in memory.

# **yield Keyword**
The yield keyword is used in generator functions to yield a value and pause execution until the next call to next().

````python
# Generator using yield
def countdown(n):
    while n > 0:
        yield n
        n -= 1

# Using the generator
for value in countdown(5):
    print(value)  # Outputs: 5, 4, 3, 2, 1
````

````python
# Generator that generates Fibonacci numbers
def fibonacci(limit):
    a, b = 0, 1
    while a < limit:
        yield a
        a, b = b, a + b

# Print Fibonacci numbers less than 20
for num in fibonacci(20):
    print(num)  # Outputs: 0, 1, 1, 2, 3, 5, 8, 13
````

* yield pauses the generator function and returns a value to the caller.
* When next() is called, execution resumes from the last yield point.
* Generators maintain the state of local variables between yield calls, unlike functions that reset their state on each call.

Methods are actions that an object can perform. They are functions defined inside a class.

````python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):  # Method
        print(f"{self.name} says Woof!")

my_dog = Dog("Buddy", "Golden Retriever")
my_dog.bark()  # Output: Buddy says Woof!
````

## **`__init__` Method**

`__init__` and `self` are key concepts in Python's object-oriented programming. They are used to initialize and manage the attributes (a variable that belongs to an object or a class) of objects in a class. Understanding these concepts helps you create reusable and organized code.

- **`__init__`:** A special method (constructor) that runs automatically when you create an object. It is used to initialize attributes.
- **`self`:** Refers to the object itself, allowing it to store and access its own unique data and methods. Using the word self is a convention. Using something else can be confusing.

# **Inheritance and Polymorphism**
These are two core concepts of Object-Oriented Programming (OOP). Let's break them down with simple definitions and examples.

### **Inheritance**
Inheritance allows a class (child class) to reuse the properties and methods of another class (parent class). It helps to reduce code repetition and build on existing functionality.

**Real-Life Analogy:**

A child inherits traits (like eye color, height) from their parent. Similarly, a child class can inherit attributes and methods from a parent class.

````python
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")

# Child class (inherits from Animal)
class Dog(Animal):
    def bark(self):
        print(f"{self.name} says Woof!")

# Creating objects
my_dog = Dog("Buddy")
my_dog.eat()  # Inherited method from Animal -> Output: Buddy is eating.
my_dog.bark()  # Method from Dog class -> Output: Buddy says Woof!
````

# **Polymorphism**

Polymorphism means "many forms." It allows methods in different classes to have the same name but behave differently, depending on the class.

**Real-Life Analogy:**

The word "run" has many forms:
A human runs on two legs.
A dog runs on four legs.
A car runs on wheels.
Even though the word is the same, the behavior differs.

````python
# Parent class
class Animal:
    def make_sound(self):
        print("Some generic animal sound.")

# Child class 1
class Dog(Animal):
    def make_sound(self):
        print("Woof!")

# Child class 2
class Cat(Animal):
    def make_sound(self):
        print("Meow!")

# Using polymorphism
def animal_sound(animal):
    animal.make_sound()  # Calls the specific version of make_sound()

dog = Dog()
cat = Cat()

animal_sound(dog)  # Output: Woof!
animal_sound(cat)  # Output: Meow!
````

# **Special Methods in Python**

Special methods (also called magic methods or dunder methods) are built-in methods in Python with names surrounded by double underscores (e `__init__`, `__str__`, `__add__`).
They let you customize how objects of your class behave in certain situations.


They make your classes act like Python's built-in types, improving usability and readability.

### **How Do They Work?**
1. Special methods are automatically called by Python when you use certain **operators** or **functions**.  
2. For example:
   - `__init__`: Called when you create an object (initializes it).  
   - `__str__`: Called when you print an object or convert it to a string.  
   - `__add__`: Called when you use `+` with objects.  

````python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):  # Customizes how the object looks as a string
        return f"'{self.title}' by {self.author}, {self.pages} pages"

    def __len__(self):  # Lets you use len() on the object
        return self.pages

# Create a Book object
book = Book("The Great Gatsby", "F. Scott Fitzgerald", 218)

# Using special methods
print(book)          # Output: 'The Great Gatsby' by F. Scott Fitzgerald, 218 pages
print(len(book))     # Output: 218
````

# **Encapsulation**

Encapsulation is about hiding the details of how something works while giving controlled access to it. In Python, this is done by keeping some data or methods private and only allowing access through specific public methods.

**Real-Life Analogy:**

Think of a TV remote:
The buttons on the remote are public (you can press them).
The internal circuits are private (you don’t need to touch or know how they work).

**Key Idea:** is to protect your data from being accessed or modified directly.

````python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  # Private attribute (using double underscores)

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid deposit amount!")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. Remaining balance: ${self.__balance}")
        else:
            print("Insufficient funds!")

# Usage
account = BankAccount("Alice", 100)
account.deposit(50)       # Deposited $50. New balance: $150
account.withdraw(30)      # Withdrew $30. Remaining balance: $120
#print(account.__balance)  # Error: Attribute is private
````

# **Decorators and Context Managers**

## **Creating and Applying Decorators**
Decorators are functions that modify the behavior of other functions or methods. They are applied using the @decorator_name syntax.

````python
# Simple decorator to log function calls
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"Calling function: {func.__name__}")
        result = func(*args, **kwargs)
        print(f"Function {func.__name__} finished execution")
        return result
    return wrapper

# Applying the decorator using @
@logger
def say_hello(name):
    print(f"Hello, {name}!")

say_hello("Alice")
````

* Decorators modify the behavior of functions or methods.
* They are defined as functions that return a "wrapper" function.
* Apply a decorator using @decorator_name before the function definition.

## **Built-in Decorators**
Python provides several built-in decorators for commonly used patterns, such as @staticmethod, @classmethod, and @property.



### **@staticmethod**
Defines a method that does not require access to the instance or class.

````python
# Static method example
class Math:
    @staticmethod
    def add(a, b):
        return a + b

result = Math.add(5, 3)  # 8
````

@staticmethod allows the method to be called on the class itself, not an instance.

### **@classmethod**
Defines a method that takes the class itself (cls) as the first argument instead of the instance.

````python
# Class method example
class Person:
    population = 0

    def __init__(self, name):
        self.name = name
        Person.population += 1

    @classmethod
    def get_population(cls):
        return cls.population

person1 = Person("Alice")
person2 = Person("Bob")

total_population = Person.get_population()  # 2
````

### **@property**
Converts a method into a read-only attribute using the @property decorator.

# **Creating Custom Context Managers**

A context manager handles resource management for you, ensuring that resources (like files or database connections) are properly opened and closed, even if an error occurs.

Sometimes, you have special tasks like setting up and cleaning up resources.
A custom context manager lets you control what happens when entering and exiting the `with` block.

A common way to create custom context managers is:




1.   Using a class with **`__enter__`:** and **`__exit__`:** methods.
2.   Using the contextlib module.



````python
class DatabaseConnection:
    def __enter__(self):
        print("Connecting to the database...")
        return self  # This object is available as `db` in the `with` block

    def __exit__(self, exc_type, exc_value, traceback): # exc_type (type of the exception),
    #exc_value (actual exception object containing the error message), traceback (the sequence of calls leading to the error: You can use the traceback object to log or analyze the exact lines of code where the error occurred.)
        print("Closing the database connection...")
        if exc_type:  # If an exception occurred
            print(f"An error occurred: {exc_value}")

# Using the custom context manager
with DatabaseConnection() as db:
    print("Performing database operations...")
    # Uncomment to test exception handling:
    # raise ValueError("Something went wrong!")
````

### **How it Works:**
1. **`__enter__`:** Called when entering the `with` block.  
   - You can initialize resources here.
   - You can return an object to use inside the block (e.g., `db`).

2. **`__exit__`:** Called when exiting the `with` block.  
   - Handles cleanup (like closing a connection).
   - Receives information about any exceptions (`exc_type`, `exc_value`,

````python
from contextlib import contextmanager

@contextmanager
def custom_manager():
    print("Entering...")
    try:
        yield "Resource"
    finally:
        print("Exiting...")

# Usage
with custom_manager() as resource:
    print(f"Using {resource}")
````

* The @contextmanager decorator makes it easier to create context managers using yield.

# **Concurrency and Parallelism**

## **Introduction to threading and multiprocessing**

Concurrency and parallelism allow multiple tasks to run simultaneously, improving performance. Python supports concurrency with threading and parallelism with multiprocessing.

### **Threading**

The threading module allows you to run **multiple** threads **concurrently**. Threads run in the same memory space and share data, making them lightweight but requiring synchronization.

````python
import threading
import time

def worker():
    print("Starting thread")
    time.sleep(2)
    print("Finished thread")

# Create and start two threads; they will run the "worker" function concurrently.
thread1 = threading.Thread(target=worker)
thread2 = threading.Thread(target=worker)

# begin the execution of "worker" in separate threads.
thread1.start()
thread2.start()

# pauses the main program until thread1 completes. same for thread2 ...
thread1.join()  # Wait for thread1 to finish
thread2.join()  # Wait for thread2 to finish
````

### **Multiprocessing**

The multiprocessing module allows you to run multiple processes in parallel. Each process runs in **its own memory space**, so data is not shared directly.



````python
import multiprocessing
import time

def worker():
    print("Starting process")
    time.sleep(2)
    print("Finished process")

# Create and start two processes.
process1 = multiprocessing.Process(target=worker)
process2 = multiprocessing.Process(target=worker)

# Begin executing the worker function in separate processes.
process1.start()
process2.start()

# Wait for Processes to Finish:
process1.join()  # Wait for process1 to finish: pauses the main program until process1 completes.
process2.join()  # Wait for process2 to finish
````

* Processes run in separate memory spaces, making them safer but more resource-intensive than threads.
* Use multiprocessing.Process(target=function_name) to create a process that runs function_name.
* process.start() begins execution of the process, and process.join() waits for it to complete.

# **functools Utilities**

The `functools` module in Python provides several utilities that make working with functions easier and more efficient. These tools help modify or enhance functions, manage caching, and work with partial functions, among other things.



### **lru_cache**

It stores the results of expensive function calls and reuses them when the same inputs occur again. This is useful for improving performance.

````python
from functools import lru_cache

@lru_cache(maxsize=3)  # Cache up to 3 recent results
def factorial(n):
    print(f"Calculating factorial({n})")
    return 1 if n == 0 else n * factorial(n - 1)

print(factorial(5))  # Calculates and caches
print(factorial(5))  # Uses the cached result
````

* @lru_cache caches function results to avoid redundant calculations.

### **partial**

Creates a new version of a function with some arguments pre-filled.

````python
from functools import partial

def power(base, exponent):
    return base ** exponent

square = partial(power, exponent=2)  # Predefine exponent as 2
print(square(5))  # Output: 25 (5^2)
````

### **reduce**
Applies a function cumulatively to the items in a sequence, reducing it to a single value.

````python
from functools import reduce

numbers = [1, 2, 3, 4]

def multiply(x, y):
    return x * y

product = reduce(multiply, numbers)
print(product)  # Output: 24 (1 * 2 * 3 * 4)
````

# **Profiling Scripts (%prun, %lprun, %memit, %mprun)**

### **%prun (Profile the execution time of a program)**

Use %prun to get detailed statistics about the execution time of each function call in your script.

````python
# Profile a function
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

%prun factorial(10)
````

* Use %prun to profile the execution of an entire function or script.
* It shows the number of calls, total time, and time per call for each function.

### **%lprun (Line-by-line profiling)**
Line-by-line profiling allows you to see how much time is spent on each line of a function. To use %lprun, you need to install the line_profiler package.

````python
# Example of line-by-line profiling (requires line_profiler)
from time import sleep

def example_function():
    sleep(1)
    result = [i ** 2 for i in range(10000)]
    sleep(2)
    return sum(result)

%lprun -f example_function example_function()
````

* Use %lprun -f function_name to profile individual lines of a function.
* Displays the time spent on each line of code in the function.
* Requires the line_profiler package (install via pip install line_profiler).

### **%memit (Memory usage measurement)**
Measure the memory usage of a function or statement using %memit. This requires the memory_profiler package.

````python
# Example of measuring memory usage
%load_ext memory_profiler

# Measure the memory usage of this list comprehension
%memit [i ** 2 for i in range(100000)]
````

* Use %memit to measure the peak memory usage of a statement.
* Requires the memory_profiler package (install via pip install memory_profiler).


### **%mprun (Memory profiling line-by-line)**
Profile memory usage line-by-line in a function. Similar to %lprun, but for memory instead of time.

````python
# Example of memory profiling line-by-line
from memory_profiler import profile

@profile
def example_function():
    x = [i for i in range(100000)]
    y = [i ** 2 for i in x]
    del x
    return y

example_function()
````

* Use %mprun -f function_name function_name() to measure memory usage of each line in a function.
* Requires the memory_profiler package (install via pip install memory_profiler).
* Must be run as a Python script (not in Jupyter) for proper profiling.

# **Understanding Data Types in Python**
Python supports various data types like `int`, `float`, and `str`. However, **NumPy** introduces specialized, efficient data types for numerical computations. NumPy arrays have fixed types, ensuring faster calculations and optimized memory usage.


**Key Points:**
- Common NumPy data types: `int32`, `float64`, `bool`.
- All elements in a NumPy array must share the same data type.

### **Creating and Manipulating Arrays**

**1. Arrays from Python Lists**
You can easily convert Python lists into NumPy arrays.

````python
import numpy as np

array_from_list = np.array([1, 2, 3, 4, 5])
print(array_from_list)
````

**2. Arrays from Scratch**
NumPy provides functions to create arrays directly without relying on lists.

````python
zeros_array = np.zeros((2, 3))      # 2x3 array of zeros
ones_array = np.ones((2, 3))        # 2x3 array of ones
random_array = np.random.rand(2, 3) # 2x3 array of random numbers
````

**3. Specifying Data Types**
Use the `dtype` parameter to control the data type of the array.

````python
int_array = np.array([1, 2, 3], dtype=np.int32)
float_array = np.array([1.2, 3.4, 5.6], dtype=np.float64)
````

### **Basics of NumPy Arrays**

**1. Attributes, Indexing, and Slicing**

Attributes like `shape`, `size`, and `dtype` help describe the array. Indexing and slicing allow access to specific elements or subarrays.

````python
array = np.array([[1, 2, 3], [4, 5, 6]])

# Attributes
print(array.shape)  # (2, 3)
print(array.size)   # 6
print(array.dtype)  # int64

# Indexing and slicing
print(array[1, 2])   # Element at row 1, column 2 -> 6
print(array[:, 1])   # All rows of column 1 -> [2, 5]
````

**2. Reshaping Arrays**

Change the shape of an array without altering its data.

````python
array = np.array([1, 2, 3, 4, 5, 6])
reshaped = array.reshape((2, 3))  # Reshaped to 2x3 array
````

**3. Array Concatenation and Splitting**

Combine or split arrays as needed.


````python
# Concatenation
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])
concatenated = np.concatenate([array1, array2])

# Splitting
split_array = np.split(array1, 3)  # Split into 3 parts
````

### **Computation on Arrays**

**1. Universal Functions (UFuncs)**

Perform operations on every element of the array.

````python
array = np.array([1, 2, 3])
squared = np.square(array)  # [1, 4, 9]
````

**2. Broadcasting**

Apply operations between arrays of different shapes or between arrays and scalars.

````python
array = np.array([1, 2, 3])
result = array + 5  # [6, 7, 8]
````

**3. Aggregations**

Compute summary statistics like sums, averages, and more.

````python
array = np.array([1, 2, 3, 4, 5])

total = np.sum(array)      # 15
minimum = np.min(array)    # 1
maximum = np.max(array)    # 5
average = np.mean(array)   # 3.0
````

### **Advanced Indexing**

**1. Fancy Indexing**
Select specific elements using a list or array of indices.

````python
array = np.array([10, 20, 30, 40, 50])
selected = array[[0, 2, 4]]  # [10, 30, 50]
````

**2. Sorting and Binning**
Sort data or organize it into bins.

````python
array = np.array([3, 1, 4, 1, 5, 9])
sorted_array = np.sort(array)  # [1, 1, 3, 4, 5, 9]
````

### **Structured Arrays**
Structured arrays allow you to work with multiple fields, similar to rows in a database.

````python
data = np.array([(25, 'Alice', 55.0), (30, 'Bob', 60.5)],
                dtype=[('age', 'i4'), ('name', 'U10'), ('weight', 'f4')])

# Access fields by name
print(data['age'])   # [25, 30]
print(data['name'])  # ['Alice', 'Bob']
````

# **Data Manipulation with Pandas**

### **Introduction to Pandas Objects**

**1. Series**  

A Series is a one-dimensional, array-like object with labeled indices. Think of it as a hybrid between a dictionary (because of labels) and a NumPy array (because of its efficiency).

````python
import pandas as pd

# Create a Series from a list
series = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print(series)
````

**Key Points:**
- A Series can store any data type (integers, floats, strings, etc.).
- You can access elements using the labels:

````python
print(series['a'])  # Output: 10
````

- Use `.values` to access the underlying data as a NumPy array and `.index` for the labels.

**2. DataFrame**

A DataFrame is a two-dimensional table-like structure with labeled rows and columns, similar to Excel spreadsheets or SQL tables.

````python
# Create a DataFrame from a dictionary
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
})
print(df)
````

**Key Points:**
- Columns are accessible like a dictionary:

````python
print(df['Name'])  # Output: A Series containing ['Alice', 'Bob', 'Charlie']
````

- Use `.shape` to find the dimensions, `.columns` for column names, and `.index` for row labels.

**3. Index**  
An Index is an immutable array of labels for rows or columns in a Series or DataFrame.

````python
# Access DataFrame index
print(df.index)  # Output: RangeIndex(start=0, stop=3, step=1)
````

**Key Points:**
- Indices provide fast label-based lookups.
- You can reset or set custom indices using `df.reset_index()` or `df.set_index()`.

Hierarchical indices allow multiple levels of labels in rows or columns, useful for multi-dimensional data.

## **Visualization with Matplotlib**

### **Basic Plots: Line, Scatter, and Bar Plots**

Matplotlib is a powerful library for creating a variety of plots to visualize data.

````python
import matplotlib.pyplot as plt

# Line Plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
plt.plot(x, y, label='Line Plot', color='blue')
plt.xlabel('X-axis (e.g., Time)')
plt.ylabel('Y-axis (e.g., Sales)')
plt.title('Line Plot Example')
plt.legend()
plt.grid(True)  # Adds a grid for better visualization
plt.show()

# Scatter Plot
plt.scatter(x, y, color='red', label='Scatter Plot', s=100)  # s sets marker size
plt.xlabel('X-axis (e.g., Age)')
plt.ylabel('Y-axis (e.g., Height)')
plt.title('Scatter Plot Example')
plt.legend()
plt.show()

# Bar Plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 20, 15, 30, 25]
plt.bar(categories, values, color='green', label='Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.legend()
plt.show()
````

- **Use Cases**:  
  - Line plots: Trends over time (e.g., stock prices).  
  - Scatter plots: Correlation between variables (e.g., height vs. weight).  
  - Bar plots: Comparing categories (e.g., survey responses).

  ### **Plot Customization: Colors, Styles, Labels, Legends**

````python
# Customized Line Plot
plt.plot(x, y, color='purple', linestyle='--', marker='o', label='Custom Line')
plt.xlabel('X-axis (Custom Example)')
plt.ylabel('Y-axis')
plt.title('Customized Line Plot')
plt.legend()
plt.grid(True)
plt.show()
````

- **Customizations**:  
  - `linestyle`: Choose styles like `'--'`, `'-.'`.

Visualize multiple plots in one figure using subplots.
